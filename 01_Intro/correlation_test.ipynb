```python
%matplotlib inline
import sympy
from sympy import symbols, Piecewise, integrate, oo
sympy.init_printing()
x, τ, t = symbols("x \\tau t", real=True)
```

```python
B = Piecewise((0, x < -0.5), (1, x < 0.5), (0, True))
sympy.plot(B, (x, -1.5, 1.5));
```

```python
B2 = integrate(B.subs(x, t) * B.subs(x, t + τ), (t, -5, 5))
```
